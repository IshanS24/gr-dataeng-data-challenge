```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 9]


df2 = pd.read_csv('data/crypto.csv')
df2
```

```python
df2
```

```python
#part 1 make a graph for 2 currencys

fig, ax= plt.subplots(figsize=(15,9))

for name, group in df2.groupby(by='Asset_Name'):
    group.plot(x='timestamp', y='Close', ax=ax, label=name)
```

```python
df2 = df2[(df2['Asset_Name']  == 'Bitcoin') | (df2['Asset_Name'] == 'Ethereum')]

fig, ax= plt.subplots(figsize=(15,9))

for name, group in df2.groupby(by='Asset_Name'):
    group.plot(x='timestamp', y='Close', ax=ax, label=name)
```
