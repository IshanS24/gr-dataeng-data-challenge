```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [15, 9]

df = pd.read_csv('data/crypto.csv')


monero = df[df['Asset_Name'] == 'Monero']
monero.plot(x='timestamp', y='Close')
```

```python

monero = monero[['Close','timestamp']].copy()
monero['change'] = monero['Close'].diff()
monero
```

```python
#already found the difference now find the highest value in the change one
max = monero['change'].max()
max
```

```python
monero['pct_change'] = monero['Close'].pct_change()
monero['pct_change'] 
```

```python
maxPercent = monero['pct_change'].max()
maxPercent
```
